```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = ("Resources/schools_complete.csv")
student_data_to_load = ("Resources/students_complete.csv")

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
df = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
schools = df['school_name'].nunique()
students = df["Student ID"].nunique()
budget = df.groupby('school_name')['budget'].mean().sum()
avgmath = df["math_score"].mean()
avgread = df["reading_score"].mean()
overallpassrate = (avgmath + avgread)/2
df_perc_pass_math = df.query("math_score >= 70") 
percent_pass_math = (df_perc_pass_math.shape[0] / df.shape[0]) * 100
# percentPass
df_perc_pass_read = df.query("reading_score >= 70")
percent_pass_read = (df_perc_pass_read.shape[0] / df.shape[0]) * 100

District_Summary=pd.DataFrame({
    "Total Schools": [schools], "Total Students": [students], "Total Budget": [budget], "Average Math Score":[avgmath],
    "Average Reading Score":[avgread], "% Passing Math": [percent_pass_math], "% Passing Reading": [percent_pass_read],
    "%Overall Passing Rate":[overallpassrate]
})
District_Summary
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)
  
* Create a dataframe to hold the above results

```python
school_type = df.groupby('school_name').max()['type']
school_student_tot = df.groupby('school_name').nunique()['Student ID']
budget_per_school = df.groupby('school_name').mean()['budget']
budget_per_student = budget_per_school/school_student_tot
school_avgmath = df.groupby('school_name').mean()['math_score']
school_avgread = df.groupby('school_name').mean()['reading_score']        
school_pass_math = df.loc[df['math_score'] >=70]
school_pass_math = school_pass_math.groupby('school_name').count()['math_score']
school_percent_pass_math = school_pass_math/school_student_tot *100
school_pass_read = df.loc[df['reading_score'] >=70]
school_pass_read = school_pass_read.groupby('school_name').count()['reading_score']
school_percent_pass_read = school_pass_read/school_student_tot *100
school_overallpassrate = (school_percent_pass_math + school_percent_pass_read)/2

School_Summary_Output=pd.DataFrame({
    "School Type": school_type, "Total Students": school_student_tot, 
    "Total School Budget": budget_per_school, "Per Student Budget": budget_per_student, 
    "Average Math Score":school_avgmath, "Average Reading Score":school_avgread, 
    "%Passing Math": school_percent_pass_math, "%Passing Reading": school_percent_pass_read,
    "%Overall Passing Rate": school_overallpassrate
})

del School_Summary_Output.index.name
School_Summary_Output
```
